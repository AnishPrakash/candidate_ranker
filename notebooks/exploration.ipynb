```python
import pandas as pd
import sys
import os

# Add the src directory to the path so we can import our ingest module
sys.path.append(os.path.abspath(os.path.join('..')))
from src.ingest import stream_candidates

# Load just the first 50 candidates for a quick structural analysis
sample_data = []
try:
    for i, cand in enumerate(stream_candidates("../candidates.jsonl")):
        sample_data.append(cand)
        if i >= 49:
            break
except FileNotFoundError:
    print("Please ensure candidates.jsonl is in the root directory.")

if sample_data:
    df = pd.json_normalize(sample_data)
    
    print(f"Loaded {len(df)} sample profiles.\n")
    print("--- Basic Demographics ---")
    print(df[['candidate_id', 'profile.years_of_experience', 'profile.current_title']].head())
    
    print("\n--- Redrob Signals Preview ---")
    signals = pd.json_normalize([c['redrob_signals'] for c in sample_data])
    print(signals[['open_to_work_flag', 'recruiter_response_rate', 'github_activity_score']].head())
    
    # Let's inspect candidate CAND_0000001 (The known keyword stuffer from the briefing)
    suspect = next((c for c in sample_data if c.get('candidate_id') == 'CAND_0000001'), None)
    if suspect:
        print("\n--- Suspect Analysis: CAND_0000001 ---")
        print(f"Claimed Title: {suspect['profile'].get('current_title')}")
        print("Claimed Skills:")
        for skill in suspect.get('skills', [])[:3]:
            print(f" - {skill.get('name')}: {skill.get('duration_months')} months")
```
